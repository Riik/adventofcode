```python
year = 2020; day = 7

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day).strip()
data = data.split("\n")
# data = [d.replace("\n", " ") for d in data]
# data = [(eval(left), eval(right)) for left, right in data]
data
```

```python
import re
```

```python
reg_holder = r"([\w ]*) bag"
reg_bag = r"([\d+]) ([\w ]*) bag"

def parse_item(item):
    m = re.match(reg_bag, item.strip())
    num, color = m.groups()
    return num, color

def parse_line(line):
    holder, items = line.split(" contain ")
    m = re.match(reg_holder, holder)
    color = m.groups()[0]
    items = items.split(",")
    if "no" in items[0]:
        return {color: None}
    items = {color: num for num, color in map(parse_item, items)}
    return {color: items}
```

```python
parse_line(data[0])
```

```python
all_rules = list(map(parse_line, data))
all_rules
```

```python
from functools import reduce
gold = "shiny gold"

def can_contain(color):
    return set([holder for rule in all_rules for holder, bags in rule.items() if bags and color in bags.keys()])

prev = set()
cur = can_contain(gold)

while len(prev) != len(cur):
    print(len(cur))
    prev = cur
    new_items = [can_contain(c) for c in cur]
    cur = reduce(lambda a, b: a | b, new_items + [prev])
    
answer = len(cur)
```

```python
answer
```

```python
submit(answer, year=year, day=day, part="a")
```

```python
rules_dict = reduce(lambda a, b: {**a, **b}, all_rules)
```

```python
rules_dict[gold]
```

```python
def expand_bag(bag_dict):
    if not bag_dict: 
        return []
    return [color for color, num in bag_dict.items() for i in range(int(num))]
```

```python
expand_bag(rules_dict[gold])
```

```python
current_bags = [gold]
bag_count = 0

    
while len(current_bags):
    current_bags = [expand_bag(rules_dict[bag]) for bag in current_bags]
    current_bags = [item for bags in current_bags for item in bags if bags]
    bag_count += len(current_bags)
    print(len(current_bags), bag_count)
    
answer = bag_count 
answer
```

```python
submit(answer, year=year, day=day, part="b")
```
