```python
year = 2020; day = 6

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)

data = data.split("\n\n")
data = [d.split("\n") for d in data]
# data = [(eval(left), eval(right)) for left, right in data]
data
```

```python
from functools import reduce
def parse_group(group):
    group = map(set, group)
    overlap = reduce(lambda a, b: a | b, group)
    return len(overlap)
```

```python
answer = sum(map(parse_group, data))
answer
```

```python
submit(answer, year=year, day=day, part="a")
```

```python
def parse_group(group):
    group = map(set, group)
    overlap = reduce(lambda a, b: a & b, group)
    return len(overlap)

answer = sum(map(parse_group, data))

submit(answer, year=year, day=day, part="b")
```
