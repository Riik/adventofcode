```python
year = 2020; day = 8

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day).strip()
# data = """nop +0
# acc +1
# jmp +4
# acc +3
# jmp -3
# acc -99
# acc +1
# jmp -4
# acc +6"""
data = data.split("\n")
data = [tuple(d.split(" ")) for d in data]
# data = [(eval(left), eval(right)) for left, right in data]
data
```

```python
def run_prog(data):
    state_hist = set()
    i = 0
    new_state = data[i]
    acc = 0
    # while new_state not in state_hist:
    for k in range(10000):
        inst, num = new_state
        print(i, inst, num)
        num = int(num)
        if inst == "jmp": 
            i += num
        if inst == "acc":
            acc += num
            i += 1
        if inst == "nop":
            i += 1
        if i in state_hist: 
            print("found repeat!")
            break
        state_hist.add(i)
        if i >= len(data):
            break
        new_state = data[i]
    else: 
        raise ValueError
    return state_hist, acc
```

```python
state_hist, acc = run_prog(data)
```

```python
submit(acc, year=year, day=day, part="a")
```

```python
import pandas as pd
prog = pd.DataFrame(data, columns=["instruction", "value"])
prog["value"] = prog["value"].astype("int")
prog["next_jmp"] = prog.index + prog["value"]
prog["next_step"] = prog.index + 1
prog.loc[prog["instruction"] == "acc", "next_jmp"] = None

prog.loc[prog["instruction"] == "acc", "valid_next"] = prog.loc[prog["instruction"] == "acc", "next_step"]
prog.loc[prog["instruction"] == "nop", "valid_next"] = prog.loc[prog["instruction"] == "nop", "next_step"]
prog.loc[prog["instruction"] == "jmp", "valid_next"] = prog.loc[prog["instruction"] == "jmp", "next_jmp"]
prog.loc[prog["instruction"] == "nop", "invalid_next"] = prog.loc[prog["instruction"] == "nop", "next_jmp"]
prog.loc[prog["instruction"] == "jmp", "invalid_next"] = prog.loc[prog["instruction"] == "jmp", "next_step"]

prog
```

```python
def get_parent(target):
    parents_valid = set(prog.index[prog["valid_next"] == target].values)    
    parents_invalid = set(prog.index[prog["invalid_next"] == target].values)    
    print(f"valid parents to {target}: {parents_valid}")
    print(f"invalid parents to {target}: {parents_invalid}")
    return parents_valid, parents_invalid

def change_to(inst):
    return "jmp" if inst == "nop" else "nop"

def find_change_instr(targets):
    for k in range(1000):
        new_targets = set()
        invalid_parents = set()
        print()
        for t in targets:
            valid_p, invalid_p = get_parent(t)
            new_targets |= valid_p
            invalid_parents |= invalid_p
        print(targets)
        if any(t in state_hist for t in invalid_parents):
            print("found something!", k)
            print(set(invalid_parents) & state_hist)
            change_instr = list(set(invalid_parents) & state_hist)[0]
            old_instr = prog.iloc[change_instr]["instruction"]
            new_instr = change_to(old_instr)
            print(f"Change instr {change_instr} from {old_instr} to {new_instr}")

            break
        targets = new_targets
    else: 
        raise ValueError("Didn't find change instr")
    return change_instr, new_instr
    
```

```python
idx, instr = find_change_instr([len(data)])
```

```python
prog.iloc[idx]
```

```python
new_data = [d if i != idx else (instr, d[1]) for i, d in enumerate(data)]
state_hist, acc = run_prog(new_data)
```

```python
submit(acc, year=year, day=day, part="b")
```
