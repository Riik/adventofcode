```python
reg_holder = r"([\w ]*) bag"
reg_bag = r"([\d+]) ([\w ]*) bag"

def parse_item(item):
    m = re.match(reg_bag, item.strip())
    num, color = m.groups()
    return num, color

def parse_line(line):
    holder, items = line.split(" contain ")
    m = re.match(reg_holder, holder)
    color = m.groups()[0]
    items = items.split(",")
    print(items)
    items = {color: num for num, color in map(parse_item, items)}
    return {color: items}
```
