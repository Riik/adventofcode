```python
year = 2020; day = 5

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)
data = data.split("\n")
# data = [d.replace("\n", " ") for d in data]
# data = [(eval(left), eval(right)) for left, right in data]
data
```

```python
test = "FBFBBFFRLR"
expected = (44, 5)
N_SEATS = 128

def parse_seat(line):
    y = "".join((map(lambda c: "0" if c == "F" else "1", line[:7])))
    y = int(y, 2)
    x = "".join((map(lambda c: "0" if c == "L" else "1", line[7:])))
    x = int(x, 2)
    return y*8+x
    
parse_seat(test)
```

```python
answer = max(parse_seat(line) for line in data)
answer
```

```python
submit(answer, year=year, day=day, part="a")
```

```python
all_ids = np.array(list(map(parse_seat, data)))
all_ids = np.sort(all_ids)
answer = all_ids[np.where(np.diff(all_ids) == 2)][0] + 1
answer
```

```python
submit(answer, year=year, day=day, part="b")
```
