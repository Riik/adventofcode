```python
year = 2020; day = 9

from aocd import get_data, submit
from aocd.models import Puzzle

puzzle = Puzzle(year, day)

import numpy as np
data = get_data(year=year, day=day).strip()
# data = puzzle.examples[0].input_data

data = data.split("\n")
data = list(map(int, data))
# data = [tuple(d.split(" ")) for d in data]
# data = [(eval(left), eval(right)) for left, right in data]
data
```

```python
from itertools import combinations

def allowed_values(hist):
    return set(x + y for x, y in combinations(hist, 2))

answer = None
window_len = 25
for i in range(window_len-1, len(data)-window_len-1):
    d = data[i:i+window_len]
    x = data[i+window_len]
    allowed = allowed_values(d)
    if not x in allowed:
        answer = x
        print(x)
        break
```

```python
submit(answer, year=year, day=day, part="a")
```
