```python
year = 2025; day = 3

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)

data = data.strip()
data = data.split("\n")
```

```python
import numpy as np
l = data[0]
ans = 0
for l in data:
    l = np.array(list(map(int, l)))
    # first digit is the highest we can get, but we need to leave at least 
    # one digit at the end
    dig1 = max(l[:-1])
    # second digit is the highest of what comes after the first one
    dig2 = max(l[min(np.where(l == dig1)[0])+1:])
    ans += int(dig1 * 10 + dig2)
ans
```

```python
submit(ans, part="a", year=year, day=day)
```

```python
import numpy as np
ans = 0
for l in data:
    digits = []
    l = np.array(list(map(int, l)))
    for i in range(11, -1, -1):
        # again first digit is the highest we can get. But we need to 
        # leave at least 11 digits after the first one, at least 10 
        # after the second one, etc. 
        new_d = max(l[:-i]) if i > 0 else max(l)
        l = l[min(np.where(l == new_d)[0])+1:]
        digits.append(new_d)
    ans += int("".join(map(str, digits)))
ans
```

```python
submit(ans, part="b", year=year, day=day)
```
