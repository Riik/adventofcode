```python
year = 2025; day = 6

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)

data = data = data.split("\n")
operands = data[-1].split()
chars = np.array([[c for c in line] for line in data[:-1]])
```

```python
from itertools import pairwise
operand_indices = list(np.where(np.array(list(data[-1])) != " ")[0])
operand_indices.append(len(data[-1]))

cols = []
for lidx, ridx in pairwise(operand_indices):
    cols.append(chars[:, lidx:ridx])
```

```python
cols[0]
```

```python
def parse_col_normal(col):
    return [int("".join(row)) for row in col]

def parse_col_vertical(block):
    nums = []
    for i in range(block.shape[1]):
        digit = "".join(block[:, i])
        if digit.strip().isdigit():
            nums.append(int(digit))
    return nums
```

```python
ans1 = 0
ans2 = 0
for block, op in zip(cols, operands):
    nums1 = parse_col_normal(block)
    nums2 = parse_col_vertical(block)
    if op == "*":
        ans1 += np.prod(nums1)
        ans2 += np.prod(nums2)
    else:
        ans1 += np.sum(nums1)
        ans2 += np.sum(nums2)
ans1 = int(ans1)
ans2 = int(ans2)
ans1, ans2
```

```python
submit(ans1, part="a", year=year, day=day)
```

```python
submit(ans2, part="b", year=year, day=day)
```
