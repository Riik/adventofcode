```python
year = 2025; day = 1

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)
data = data.strip()
data = data.split("\n")
```

```python
val = 50
ans = 0
for l in data:
    if l[0] == "L":
        val -= int(l[1:])
    else:
        val += int(l[1:])
    if val % 100 == 0:
        ans += 1
ans
```

```python
submit(ans, part="a", year=year, day=day)
```

```python
# made this one work with the test cases below but if feels a bit disgusting
def count_crossings(prev, val):
    ans = 0
    if val == 0:
        return 1
    if prev < 0 and val >= 0:
        ans += 1
    if prev == 0 and val <= 0:
        ans -= 1
    if val < 0:
        val -= 1
    ans += abs(val // 100)
    return ans
```
