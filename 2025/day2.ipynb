```python
def check_valid_id(id_num):
    if len(id_num) % 2 != 0:
        return True
    else:
        return id_num[:len(id_num)//2] != id_num[len(id_num)//2:]
```

```python
def check_repeating_part(repeat_count, left, right):
    invalid_sum = 0
    check_length = len(left)
    start_range = int(left[:check_length // 2])
    end_range = int(right[:check_length // 2]) + 1
    # print(start_range, end_range)
    for left_part in range(start_range, end_range):
        potential_id = int(str(left_part)*repeat_count)
        # print(f"{potential_id=}")
        if potential_id <= int(right) and potential_id >= int(left):
            # print(potential_id)
            invalid_sum += potential_id
    return invalid_sum
```

```python
invalid_sum = 0
for line in data:
    left, right = line.split("-")
    # print(f"----- {left} - {right}")
    assert abs(len(left) - len(right)) <= 1
    check_length = len(left) if len(left) % 2 == 0 else len(right)
    if check_length % 2 != 0:
        continue
    left = left if len(left)%2 == 0 else left.zfill(check_length)
    right = right if len(right)%2 == 0 else "9"*check_length
    assert len(left) == len(right)
    repeating_part = left[:len(left)//2]
   
   
    invalid_sum += check_repeating_part(2, left, right)
    
invalid_sum
```
