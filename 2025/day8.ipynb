```python
year = 2025; day = 8

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)
data = [line.split(",") for line in data.split("\n")]
data = np.array(data).astype("int")
```

```python
# (n, 1, 3) - (1, n, 3) will result in (n, n, 3)
# taking the norm over the last axis will make it 2D (n, n) distance matrix
# learned this from AI while cleaning up the code after first submission
# Used double for loop first which took about 2s for 1000 points
diff = data[:, np.newaxis, :] - data[np.newaxis, :, :]
dist = np.linalg.norm(diff, axis=2)
```

```python
# np.unique also sorts 
close = np.unique(dist[dist > 0.])

# mark all points below the threshold
threshold = close[1000]
adj = (dist <= threshold) & (dist > 0)
```

```python
import networkx as nx
G = nx.from_numpy_array(adj)
components = list(nx.connected_components(G))
```

```python
lens = np.sort([len(c) for c in components])
ans1 = int(np.prod(lens[-3:]))
```

```python
submit(ans1, part="a", day=day, year=year)
```

```python
# adj = np.zeros(dist.shape)
G = nx.from_numpy_array(adj)
components = list(nx.connected_components(G))

# loop over all distances in upper triangle (dist is symmetric)
# and zip dist cost with coordinates
# didn't know about this triu_indices but learned it from AI afterwards cleaning up
triu_indices = np.triu_indices_from(dist, k=1)
edges = list(zip(triu_indices[0], triu_indices[1], dist[triu_indices]))
# order in which to add edges
edges_sorted = sorted(edges, key=lambda x: x[2])

for idx, edge in enumerate(edges_sorted[1000:]):
    i, j, _ = edge
    G.add_edge(i, j)
    # time to get a little bit emperical
    if idx > 6800:
        num_components = nx.number_connected_components(G)
        if num_components == 1: 
            final_i, final_j = i, j
            break
```

```python
ans2 = int(data[final_i][0] * data[final_j][0])
ans2
```

```python
submit(ans2, part="b", day=day, year=year)
```
