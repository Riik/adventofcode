```python
dist = np.zeros((data.shape[0], data.shape[0]))
for i, a in enumerate(data):
    for j, b in enumerate(data): 
        dist[i, j] = np.linalg.norm(a - b)
```

```python
close = np.unique(np.sort(dist[dist > 0.]))
```

```python
adj = np.zeros(dist.shape)

for i in range(1000):
    d = close[i]
    locs = np.where(dist == d)
    assert len(locs) == 2
    adj[locs] = 1
```

```python
import networkx as nx
G = nx.from_numpy_array(adj)
components = list(nx.connected_components(G))
```

```python
lens = np.sort([len(c) for c in components])
lens
```
