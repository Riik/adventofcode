```python
from dataclasses import dataclass

# I wondered why my code kept hanging so ended up implementing my own range() class
# turns out min(range()) is very inefficient in Python and you should just take range.start
# lesson learned I guess, but I'll keep this her as a reminder
@dataclass
class IdRange:
    start: int
    stop: int

    def __contains__(self, item):
        return item >= self.start and item < self.stop
```

```python
year = 2025; day = 5

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)

data = data.strip()
ranges, ids = data.split("\n\n")

ids = list(map(int, ids.split("\n")))
ranges = ranges.split("\n")
ranges = [list(map(int, r.split("-"))) for r in ranges]
ranges = [range(r[0], r[1]+1) for r in ranges]
```

```python
fresh_ids = set()
for idx in ids:
    for r in ranges:
        if idx in r:
            fresh_ids.add(idx)
            continue
ans = len(fresh_ids)
```

```python
submit(ans, year=year, day=day, part="a")
```

```python
# straight up copied those functions from 2022 day 15. Only had to fix the min() in the sort
def merge_ranges(left, right):
    if left.stop >= right.start:
        new_min = min(left.start, left.start)
        new_max = max(left.stop, right.stop)
        return [range(new_min, new_max)]
    return [left, right]

def overlap_ranges(ranges):
    ranges.sort(key=lambda r: r.start)
    output = []
    l = ranges[0]
    
    for r in ranges[1:]:
        res = merge_ranges(l, r)
        if len(res) == 2:
            l = res[1]
            output.append(res[0])
        else: 
            l = res[0]
    output.append(l)
    return output

matched_ranges = overlap_ranges(ranges)
```

```python
ans = 0
for r in matched_ranges:
    ans += r.stop - r.start
ans
```

```python
submit(ans, year=year, day=day, part="b")
```
