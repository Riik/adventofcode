```python
year = 2025; day = 7

from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)
from helpers.helpers import parse_single_matrix

data = parse_single_matrix(data)
H, W = data.shape
```

```python
tachs = {np.where(data[0, :] == "S")[0][0]}
splits = 0
for y in range(1, len(data)): 
    new_tachs = set()
    for x in tachs: 
        if data[y, x] == "^":
            splits += 1
            if x-1 > 0:
                new_tachs.add(x-1)
            if x+1 <= W:
                new_tachs.add(x+1)
        else:
            new_tachs.add(x)
    tachs = new_tachs
splits
```

```python
from collections import defaultdict
tachs = {
    np.where(data[0, :] == "S")[0][0]: 1
    }
splits = 0
for y in range(1, len(data)): 
    new_tachs = defaultdict(int)
    for x, count in tachs.items(): 
        if data[y, x] == "^":
            splits += 1
            if x-1 >= 0:
                new_tachs[x-1] += count
            if x+1 < W:
                new_tachs[x+1] += count
        else:
            new_tachs[x] += count
    tachs = new_tachs
ans1 = splits
ans2 = sum(tachs.values())
ans1, ans2
```

```python
submit(ans1, part="a", year=year, day=day)
```

```python
submit(ans2, part="b", year=year, day=day)
```
