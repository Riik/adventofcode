```python
with open("input.txt") as f:
    data = f.read().splitlines()
    data = [l.split("-") for l in data]
data
```

```python
data[:]
```

```python
nodes = list(set([n for link in data for n in link]))
```

```python
def make_links(data):
    links = {}
    for a,b in data:
        if a in links:
            links[a].append(b)
        else:
            links[a] = [b]
        if b in links:
            links[b].append(a)
        else:
            links[b] = [a]
    return links
```

```python
links = make_links(data)
```

```python
for n in nodes:
    if n not in links: 
        links[n] = []
links
```

```python
visits = dict(zip(nodes, [0]*len(nodes)))
visits['start'] += 1
visits
```

```python
def make_path(visits, node, path, can_double_visit):
    if node == "end":
        path.append(can_double_visit)
        return [path]
    paths = []
    neighbours = links[node]
    candidates = [n for n in neighbours 
                  if n.isupper()
                  or (visits[n] < 1)
                  or (visits[n] < 2 and can_double_visit)
                  and n != 'start']
    if candidates:
        for c in candidates:
            v = visits.copy()
            v[c] += 1
            d = can_double_visit if (v[c] == 1 or c.isupper()) else False
            p = path.copy()
            p.append(c)
            paths.extend(make_path(v, c, p, d))
        return paths
    return []
```

```python
all_paths = make_path(visits, 'start', ['start'], True)
sorted(all_paths)
```
