```python
def calc_magnitude(pair):
    a, _, b, = pair
    output = [str(3*int(a)+2*int(b))]
    return output
```

```python
digits = list(map(str, range(10)))
def add_explode(direction, value, data):
    if direction == -1:
        data = list(reversed(data))
    for p in range(len(data)):
        if data[p] in digits:
            data = data[:p] + [str(int(data[p])+value)] + data[p+1:]
            break
    if direction == -1:
        data = list(reversed(data))
    return data

    
assert add_explode(1, 2, list("],1],2],3],4]")) == list("],3],2],3],4]")
assert add_explode(-1, 2, list("[[[[[")) == list("[[[[[")
assert add_explode(-1, 3, list("[7,[6,[5,[4,")) == list("[7,[6,[5,[7,")
```

```python
data = "[[[[[9,8],1],2],3],4]"

def explode_once(data):
    depth = 0
    N = len(data)
    p = 0 
    popen = []
    for p in range(N):
        if data[p] == '[':
            popen.append(p)
            depth += 1
        if data[p] == ']':
            po = popen.pop()
            if depth == 5:
                datastr = "".join(data[po:p+1])
                a = int(data[po+1])
                b = int(data[p-1])
                data[:po] = add_explode(-1, a, data[:po])
                data[p+1:] = add_explode(1, b, data[p+1:])
                data[po:p+1] = '0'
                break
            else:
                depth -= 1
    else:
        return None
    return data
  
assert explode_once(list("[[[[[9,8],1],2],3],4]")) == list("[[[[0,9],2],3],4]")
assert explode_once(list("[7,[6,[5,[4,[3,2]]]]]")) == list("[7,[6,[5,[7,0]]]]")
assert explode_once(list("[[6,[5,[4,[3,2]]]],1]")) == list("[[6,[5,[7,0]]],3]")
assert explode_once(list("[[3,[2,[1,[7,3]]]],[6,[5,[4,[3,2]]]]]")) == list("[[3,[2,[8,0]]],[9,[5,[4,[3,2]]]]]")
assert explode_once(list("[[3,[2,[8,0]]],[9,[5,[4,[3,2]]]]]")) == list("[[3,[2,[8,0]]],[9,[5,[7,0]]]]")
assert explode_once(list("[[[[0,9],2],3],4]")) == None
```

```python
import math
def split_once(data):
    depth = 0
    N = len(data)
    p = 0 
    popen = []
    for p in range(N):
        if len(data[p]) > 1:
            x = int(data[p])
            split_insert = list(f"[{math.floor(x/2)},{math.ceil(x/2)}]")
            data[p:p+1] = split_insert 
            break
    else:
        return None
    return data

assert split_once(['[', '1', ',', '10', ']']) == list("[1,[5,5]]")
assert split_once(['[', '11', ',', '9', ']']) == list("[[5,6],9]")
assert split_once(['[', '12', ',', '1', ']']) == list("[[6,6],1]")
assert split_once(list("[1,[2,3]]")) == None
```

```python
data = "[[[[8,7],[7,7]],[[8,6],[7,7]]],[[[0,7],[6,6]],[8,7]]]"

def parse_magnitude(data):
    depth = 0
    data = list(data)
    N = len(data)
    p = 0 
    popen = []
    while p < N:
        if data[p] == '[':
            popen.append(p)
            depth += 1
        if data[p] == ']':
            po = popen.pop()
            value = calc_magnitude(data[po+1:p])
            data = data[:po] + value + data[p+1:]
            if len(data) < 3:
                break
            depth -= 1
            p = 0 
            popen = []
            continue
        p += 1
    return int(data[0])
    
```

```python
from functools import reduce
def compute_sum(numbers):
    return reduce(add_numbers, numbers[1:], numbers[0])
```
