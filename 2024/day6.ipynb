```python
from helpers.helpers import *

year = 2024; day = 6
```

```python
from aocd import get_data, submit
import numpy as np
data = get_data(year=year, day=day)

data = data.strip()
data = parse_single_matrix(data)

start_pos = np.where(data == "^")
y, x = int(start_pos[0][0]), int(start_pos[1][0])
data[y, x] = "."
print(y, x)

data.shape
```

```python
y, x = int(start_pos[0][0]), int(start_pos[1][0])
dy, dx = -1, 0

visited  = {(y,x)}
for _ in range(100_000):
    if y+dy not in range(data.shape[0]) or x+dx not in range(data.shape[1]):
        print("done")
        break

    if data[y+dy, x+dx] == "#":
        dy, dx = dx, -dy
    else:
        y += dy
        x += dx

    visited.add((y, x))
else:
    ValueError("out of iterations")

len(visited)
```

```python
submit(len(visited), part="a", year=year, day=day)
```

```python
wall_positions = set()

def search_pattern(d, block_y, block_x):
    y, x = int(start_pos[0][0]), int(start_pos[1][0])
    dy, dx = -1, 0

    if (y, x) == (block_y, block_x):
        return 0

    visited_with_dir = {(y,x, dy, dx)}

    for _ in range(10_000):
        if y+dy not in range(d.shape[0]) or x+dx not in range(d.shape[1]):
            return 0

        if d[y+dy, x+dx] == "#" or (y+dy, x+dx) == (block_y, block_x):
            dy, dx = dx, -dy
        else:
            y += dy
            x += dx

        if (y, x, dy, dx,) in visited_with_dir:
            return 1

        visited_with_dir.add((y, x, dy, dx,))
    raise ValueError

for y in range(data.shape[0]):
    if y % 10 == 0:
        print(y)
    for x in range(data.shape[1]):
        if data[y, x] == "#":
            continue
        if search_pattern(data, y, x):
            wall_positions.add((y, x))

wall_positions
```

```python
submit(len(wall_positions), part="b", year=year, day=day)
```
